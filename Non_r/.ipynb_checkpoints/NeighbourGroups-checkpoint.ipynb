```python
import sys
import numpy as np
import pandas as pd
#import Pyreadstat
import seaborn as sns
import matplotlib.pyplot as plt
from ete3 import ClusterTree
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
# https://stackoverflow.com/questions/31033835/newick-tree-representation-to-scipy-cluster-hierarchy-linkage-matrix-format

def newick_to_linkage(newick: str) -> (np.ndarray, [str]):
    """
    Convert newick tree into scipy linkage matrix

    :param newick: newick string, e.g. '(A:0.1,B:0.2,(C:0.3,D:0.4):0.5);'
    :returns: linkage matrix and list of labels
    """
    # newick string -> cophenetic_matrix
    tree = ClusterTree(newick)
    cophenetic_matrix, newick_labels = tree.cophenetic_matrix()
    cophenetic_matrix = pd.DataFrame(
        cophenetic_matrix, columns=newick_labels, index=newick_labels)

    # reduce square distance matrix to condensed distance matrices
    pairwise_distances = pdist(cophenetic_matrix)
    
    # return linkage matrix and labels
    return linkage(pairwise_distances), list(cophenetic_matrix.columns)


def readNewick(nwk):
    with open(nwk) as fh:
        return fh.readline()
```

```python
nwk = readNewick('../../CC353_Analysis/10359_grapetree/BIGSdb_024808_1423149961_99700_tree.nwk')
```

```python
linkageMatrix, labels = newick_to_linkage(newick=nwk)
```

```python
all_cluster = []
for a in range(2,40+1):
    clusters = fcluster(linkageMatrix, t=a, criterion='maxclust', depth=2, R=None, monocrit=None)
    labelsID = pd.DataFrame(clusters, labels, columns=[f'{a}-NighborGroup'])
    all_cluster.append(labelsID)
all_cluster = pd.concat(all_cluster, axis=1)
```

```python
# How many isolates in each cluster? This is very imbalanced as you can see! i.e many have only 1 isolate
vcs = all_cluster.apply(pd.value_counts)
```

```python
vcs = vcs.divide(vcs.sum())
```

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.heatmap(vcs.T, cmap='viridis', ax=ax)
fig.savefig('../Figures/NeighbourGroups_size_comparison.pdf')
```

```python
vcs = pd.concat([vcs[c].sort_values(ascending=False).reset_index(drop=True) for c in vcs.columns], axis=1)
```

```python
vcs = vcs.cumsum()
```

```python
import catboost
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
labelledData.head()
```

```python
target = 'Group'

X = labelledData[['aspA', 'glnA', 'gltA', 'glyA', 'pgm', 'tkt', 'uncA', 'Group']].copy().astype('category')
y = X.pop(target)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, train_size=0.8, test_size=0.2)
```

```python
#If you want to run the model this is the one
model = CatBoostClassifier(verbose=0)
model = model.fit(X_train, y_train, cat_features=list(X.columns))
```

```python
y_pred = model.predict(X_test)
labelledData['Prediction_Catboost'] = model.predict(X)
labelledData.to_excel("labelledCatBoost.xlsx") 
```
