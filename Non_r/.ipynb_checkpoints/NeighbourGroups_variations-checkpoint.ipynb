```python
import os
import sys
import glob
import numpy as np
import pandas as pd
import catboost
import seaborn as sns
from pathlib import Path
from itertools import combinations
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
def getAllCombinations(arr: list):
    """ Return all combinations, of all length of x """
    allCombinations = []
    for i in range(1, len(arr)+1):
        els = [list(x) for x in combinations(arr, i)]
        allCombinations.extend(els)
    return allCombinations
```

```python
labelledData = pd.read_excel("../../CC353_Analysis/labelledData.xlsx")
labelledData.head()
```

```python
Data_coreMLST = (
    pd.read_csv("../../Data/10359_coreMLST_dataframe.txt",sep='\t')
    .fillna(-1).rename({'id': 'ID'}, axis=1)
    .drop(['cgST (C. jejuni / C. coli cgMLST v1.0)'], axis=1)
)

for col in Data_coreMLST.columns:
    Data_coreMLST[col] = Data_coreMLST[col].apply(lambda x: -1 if ';' in str(x) else x)
Data_coreMLST = Data_coreMLST.astype(int)
```

```python
labelledData = pd.merge(labelledData, Data_coreMLST, left_on='ID', right_on='ID')
labelledData.to_pickle('../../Data/labelledData-allLoci.pkl')
```

```python
labelledData
```

```python
# Write all loci names for cluster submission
with open('../../Data/allLociNames.txt', 'w') as fh:
    for col in Data_coreMLST.columns:
        fh.write(f'{col}\n')
```

```python
target = 'Group'
loci = ['aspA', 'glnA', 'gltA', 'glyA', 'pgm', 'tkt', 'uncA']
```

```python
all_loci = getAllCombinations(loci)
len(all_loci)
```

```python
out = '../../CC353_Analysis/NG_variation'
```

```python
for locus in all_loci:
    name = "-".join(locus)
    file = f'{out}/{name}.txt'
    print(file)
    if os.path.exists(file) and os.path.getsize(file) > 0:
        continue
    with open(file, 'w') as fh:
        X = labelledData[locus + [target]].copy().astype('category')
        y = X.pop(target)

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, random_state=42, train_size=0.8, test_size=0.2)

        #If you want to run the model this is the one
        model = CatBoostClassifier(verbose=0)
        model = model.fit(X_train, y_train, cat_features=list(X.columns))

        y_pred = model.predict(X_test)
        score = adjusted_rand_score(y_test, y_pred.flatten())
        fh.write(f'{name}\t{score}\n')
```

```python
df = []
for file in glob.glob(f'{out}/*txt'):
    with open(file) as fh:
        data = fh.readline().split('\t')
    group = tuple(data[0].split('-'))
    score = float(data[1])
    n = len(group)
    df.append([group, score, n])
df = pd.DataFrame(df, columns=['loci', 'score', 'n'])
```

```python
meanScore = []
for locus in loci:
    sub = df.loc[df['loci'].apply(lambda x: locus in x)]
    meanScore.append(sub.groupby('n')['score'].mean().rename(locus))
meanScore = (
    pd.concat(meanScore, axis=1)
    .reset_index()
    .melt(id_vars='n', var_name='locus', value_name='score')
)
```

```python
MLST_7_assignment_score = sns.lineplot(x='n', y='score', hue='locus', data=meanScore)
```
