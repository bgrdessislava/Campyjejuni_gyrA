```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
group1 = '/Users/user/Documents/OneDrive - Nexus365/PhD/Campy_Analysis_ALL/Data/Group1_genomecomparator_result.xlsx'
group1 = pd.read_excel(group1).drop(['Product', 'Full name'], axis=1).set_index('Locus').T
```

```python
group1
```

```python
group2 = '/Users/user/Documents/OneDrive - Nexus365/PhD/Campy_Analysis_ALL/Data/Group2_genomecomparator_result.xlsx'
group2 = pd.read_excel(group2).drop(['Product', 'Full name'], axis=1).set_index('Locus').T
```

```python
for rmlst in group1.columns: 
    labels = list(set(list(group1[rmlst]) + list(group2[rmlst])))

    data1 = group1[rmlst].value_counts().reindex(labels).fillna(0)
    data2 = group2[rmlst].value_counts().reindex(labels).fillna(0)

    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1 = ax1.pie(data1, labels=labels)
    ax2 = ax2.pie(data2, labels=labels)
    fig.tight_layout()
    fig.savefig(f'/Users/user/Documents/OneDrive - Nexus365/PhD/Campy_Analysis_ALL/CC353_Analysis/Piechart/{rmlst}.png')

```
