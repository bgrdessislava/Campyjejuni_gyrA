```python
sub = df_raw[['ID', 'PRODUCT', 'Year']].drop_duplicates().dropna().copy()
sub = sub.pivot(index='ID', columns='PRODUCT', values='Year')
cols = sub.count().sort_values(ascending=False).index
index = sub.count(axis=1).sort_values(ascending=False).index
sub = sub.reindex(index)[cols]

fig, ax = plt.subplots(figsize=(12,6))
sns.heatmap(sub, cmap='viridis', ax=ax)
fig.tight_layout()
```
